- 3) Se desarrolla el método de mínimos cuadrados para calcular el valor de la resistencia $R$.

Cuando se utiliza el método de mínimos cuadrados es necesario encontrar de la ecuación lineal

$$
y=mx+b
$$

los parámetros $m$ y $b$, es decir, la pendiente de la recta $m$ y el punto de corte con el eje $y$, respectivamente.

En este caso el eje $x$ será el de las corrientes y el eje $y$ el de los voltajes, es decir:

$$
V= R I + b
$$

La pendiente de la recta será el valor de la resistencia, $m=R$.

Por el método de mínimos cuadrados tenemos que:

\begin{equation}
R =\dfrac{ n\sum\limits_{i=1}^n I_i V_i-\sum\limits_{i=1}^n I_i \sum\limits_{i=1}^n V_i}{n \sum\limits_{i=1}^n I_i^2-\left(\sum\limits_{i=1}^n I_i\right)^2} \tag{1}
\end{equation}

\begin{equation}
b = \dfrac{\sum\limits_{i=1}^n I_i^2 \sum\limits_{i=1}^n V_i-\sum\limits_{i=1}^n I_i \sum\limits_{i=1}^n I_i V_i}{n\sum\limits_{i=1}^n I_i^2-\left(\sum\limits_{i=1}^n I_i\right)^2} \tag{2}
\end{equation}

Primero haremos todos los cálculos de manera manual utilizando una calculadora. Por lo tanto, es necesario encontrar el valor numérico para las siguientes series de sumas:

$$
\sum V_i,\, \quad \sum I_i,\,\quad \sum I_i^2 \quad \text{y} \quad  \sum I_i V_i
$$

Con los datos dados estas sumas resultan ser:

$$
\sum V_i=0,11600 \,, \quad \sum I_i= −0,00800  \,, \quad \sum I_i^2= 0,03491 \,, \quad \sum I_i V_i= 5,25194
$$

Con estos resultados se obtienen los siguientes valores al sustituir en la ecuación (1) para $R$ y en la ecuación (2) para $b$

$$
R= 150,449 \, \Omega \, \quad \text{y}\quad \,  b= 0,066 V.
$$

Ahora es necesario calcular  el error estándar residual y los errores estándar de la pendiente $\Delta R$ y el término independiente $\Delta b$ de la regresión lineal. Pueden consultar el capítulo 5 de la [Guía General para Laboratorios](https://github.com/hectorfro/CursosUIS/blob/main/LabFisII24B/LabFisV30.pdf)

Procedemos a calcular los erroes:

\begin{equation}
\Delta R =\frac{\hat{\sigma}^2}{\sum_{i=1}^n\left(x_i-\bar{x}\right)^2} \tag{3}
\end{equation}
\begin{equation}
\Delta b = \hat{\sigma}^2\left[\frac{1}{n}+\frac{\bar{x}^2}{\sum_{i=1}^n\left(x_i-\bar{x}\right)^2}\right]  \tag{4}
\end{equation}
donde

\begin{equation}
\hat{\sigma}^2=\frac{{\mathrm{SSR}}}{n-2} =
\frac{\sum\limits_{i=1}^n \left(V_i-b-R I_i\right)^2}{n-2} \tag{5}
\end{equation}

Procedemos a calcular el numerador de la ecuación (5) con nuestros datos
$$
\begin{array}{c|c}
n& \sum\limits_{i=1}^n \left(V_i-b-R I_i\right)^2  \\ \hline\hline
20 & 0,23873
\end{array}
$$

Sustituyendo los valores numéricos se obtiene

$$
\hat{\sigma}^2= 0,11516 \,, \qquad
\Delta R=0,61638 \, \Omega \,, \qquad
\Delta b= 0,025752 \, V
$$

Por lo tanto, por el método de mínimos cuadrados se obtiene un valor para la resistencia de
$$
R= 150,4 \pm 0,6 \, \Omega
$$
Lo que representa un error porcentual del 0,4 %.


El valor obtenido para $b$ y redondeando a dos decimales es: $b= 0,07 \pm 0,03$ V.


Este resultado se puede comparar con el que se obtiene usando la librería de Numpy <<linalg.lstsq>>. Con esta librería numérica se obtiene el mismo resultado: $(m=150.449,b= 0.066)$

# Analisis de los datos

```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
# Los datos de voltajes Vn y corrientes In
Vn= array([10.120, 9.060, 8.100, 7.120, 6.140, 5.110, 4.050, 3.064,
           2.137, 1.090, -10.080, -9.050, -8.050, -7.120, -6.110,
           -5.090, -4.120, -3.091, -2.058, -1.106])
In= array([0.067, 0.060, 0.054, 0.046, 0.041, 0.033, 0.027, 0.020,
           0.013, 0.005,-0.067, -0.060, -0.055, -0.048, -0.041,
           -0.035, -0.027, -0.020, -0.014, -0.007])
```

```python
# El número de datos lo llamaremos n
n=len(Vn)
n
```

```python
# Para mejorar la visualización de expresiones usamos:
from IPython.display import display, Math
# Las sumatorias necesarias son:
Sum_I = sum(In)
Sum_V = sum(Vn)
Sum_II = sum(In**2)
Sum_IV = sum(In*Vn)
# Formateamos para 5 decimales cada expresión
expre = (
    r'n = {}, \sum I_n ={:.5f}, \sum V_n ={:.5f}, '
    r'\sum I_n^2 ={:.5f}, \sum (I_n \cdot V_n) ={:.5f}'
    .format(n, Sum_I, Sum_V, Sum_II, Sum_IV))
# Mostramos las expresiones formateadas con LaTeX
display(Math(expre))
```

```python
# El cálculo de b y m
b=(Sum_II*Sum_V-Sum_IV*Sum_I)/(n*Sum_II-Sum_I**2)
m=(n*Sum_IV-Sum_I*Sum_V)/(n*Sum_II-Sum_I**2)
print('m=',m.round(3), ',', 'b=',b.round(4))
```

```python
# Hacemos la gráfica V vs I
# Datos
x1=In
y1=Vn
# linea recta
y2= m*In+b
#
plt.figure()
plt.scatter(x1, y1, color='b',marker='+', label='datos medidos')
plt.plot(In,y2 , color="red", linewidth=+1,
         label= f'$V =$ {m:.2f} $I$ + {b:.2f}')
plt.grid(linestyle='dotted')
plt.legend(loc='best')
plt.title(r'Voltaje vs Corriente', fontsize=18)
plt.xlabel(r'$I$ [A]', fontsize=16)
plt.ylabel(r'$V$ [V]', fontsize=16)
plt.show()
```

# Cálculo de los errores

Es recomendable consultar el capítulo 5 de la [Guía General para Laboratorios](https://github.com/hectorfro/CursosUIS/blob/main/LabFisII24B/LabFisV30.pdf)

```python
# Calculamos lasuma residual de cuadrados
SSR= sum((Vn - b-m*In)**2)
# Calculamos la Desviación estándar
s_hat= sqrt(SSR/(n-2))
print('n=',n,',','SSR=', SSR ,',', 'sigma=',s_hat)
```

```python
# Los promedios de In y Vn son
Ip= sum(In)/n
Vp= sum(Vn)/n
```

```python
# Calculamos ∆m y ∆b
Delta_m = sqrt(s_hat**2 / sum((In - Ip)**2))
Delta_b= sqrt(s_hat**2 * (1/n + Ip**2 / sum((In - Ip)**2)))
```

```python
print(f'm = {round(m, 1)} \u00B1 {round(Delta_m, 1)}')
print(f'b = {round(b, 2)} \u00B1 {round(Delta_b, 2)}')
```

# Utilizando la librería "statsmodels"

También podemos usar una librería llamada Statsmodels, que es una librería de Python diseñada para realizar análisis estadístico en profundidad. Es ampliamente utilizada en campos como la econometría, la biología, la psicología, y otras disciplinas que requieren modelado estadístico y pruebas de hipótesis rigurosas.

Características Principales:

	1.	Modelos Estadísticos: Statsmodels proporciona una amplia gama de modelos estadísticos, incluyendo regresiones lineales y no lineales, modelos de series temporales (ARIMA, SARIMA), modelos de efectos fijos y aleatorios, modelos de regresión logística, y mucho más.
	2.	Pruebas Estadísticas: La librería ofrece una variedad de pruebas estadísticas, como la prueba t, la prueba de chi-cuadrado, y la prueba de normalidad de Shapiro-Wilk, entre otras.
	3.	Estimación de Parámetros: Statsmodels permite estimar parámetros de modelos estadísticos utilizando métodos como la máxima verosimilitud y los mínimos cuadrados ordinarios (OLS).
	4.	Análisis de Residuos y Diagnósticos: Una de las características destacadas de Statsmodels es su capacidad para realizar diagnósticos completos de modelos, incluyendo análisis de residuos, detección de multicolinealidad, y pruebas de heterocedasticidad.
	5.	Integración con Pandas: Statsmodels se integra perfectamente con Pandas, lo que facilita el manejo de datos y la extracción de información estadística directamente desde estructuras de datos familiares.
	6.	Interfaz de Usuario Completa: A través de una interfaz clara y orientada a objetos, Statsmodels permite a los usuarios definir modelos, ajustarlos a datos, y generar resúmenes detallados, gráficos de diagnóstico, y predicciones.

Es también recomendable consultar el capítulo 5 de la [Guía General para Laboratorios](https://github.com/hectorfro/CursosUIS/blob/main/LabFisII24B/LabFisV30.pdf)


```python
# Importamos la librería
import statsmodels.api as sm
```

```python
# Agregamos una constante (columna de unos) a xn
X = sm.add_constant(In)
# Ajustamos el modelo
model = sm.OLS(Vn, X).fit()
# Obtenemos los coeficientes y los errores estándar
b, m = model.params
Delta_b , Delta_m = model.bse
```

```python
# Imprimimos los coeficientes y sus errores estándar
print(f'm = {m:.4f} \u00B1 {Delta_m:.4f}')
print(f'b = {b:.4f} \u00B1 {Delta_b:.4f}')
```

```python
# Imprimimos los coeficientes con las cifras significativas correctas
print(f'm = {m:.1f} \u00B1 {Delta_m:.1f}')
print(f'b = {b:.2f} \u00B1 {Delta_b:.2f}')
```
