- 3) Se desarrolla el método de mínimos cuadrados para calcular $R$.

Cuando se utiliza el método de mínimos cuadrados es necesario encontrar de la ecuación lineal

$$
y=mx+b
$$

los parámetros $m$ y $b$, es decir, la pendiente de la recta $m$ y el punto de corte con el eje $y$, respectivamente.

En este caso el eje $x$ será el eje de las corrientes y el eje $y$ el de los voltajes:

$$
V= R I + b
$$

Es decir, la pendiente de la recta es el valor de la resistencia, por lo tanto:

$$
R =\dfrac{ n\sum\limits_{i=1}^n I_i V_i-\sum\limits_{i=1}^n I_i \sum\limits_{i=1}^n V_i}{n \sum\limits_{i=1}^n I_i^2-\left(\sum\limits_{i=1}^n I_i\right)^2}\,,\qquad
b = \dfrac{\sum\limits_{i=1}^n I_i^2 \sum\limits_{i=1}^n V_i-\sum\limits_{i=1}^n I_i \sum\limits_{i=1}^n I_i V_i}{n\sum\limits_{i=1}^n I_i^2-\left(\sum\limits_{i=1}^n I_i\right)^2}
$$

Primero es necesario encontrar la siguiente serie de sumas:

$$
\sum V_i,\, \sum I_i,\, \sum I_i^2 \, \text{y} \, \sum I_i V_i
$$.

Estas sumas son:

$$
\sum V_i=0.1160 \,, \quad \sum I_i= -0.0080  \,, \quad \sum I_i^2= 0.034912 \,, \quad \sum I_i V_i= 5.2520
$$

Con estos resultados se obtienen los siguientes valores

$$
R= 150.449 \, \Omega \, \text{y} \,  b= 0.066 V.
$$

Es necesario calcular los errores absolutos $\Delta R$ y $\Delta b$

$$
\Delta R  =\left[\frac{n}{n \sum\limits_{i=1}^n\left(x_i\right)^2-\left(\sum\limits_{i=1}^n x_i\right)^2}\right]^{\frac{1}{2}} S_y \,,\qquad
\Delta b  =\left[\frac{\sum\limits_{i=1}^n\left(x_i\right)^2}{n \sum\limits_{i=1}^n\left(x_i\right)^2-\left(\sum\limits_{i=1}^n x_i\right)^2}\right]^{\frac{1}{2}} S_y
$$

donde

\begin{equation}
S_y=\left[\frac{\sum\limits_{i=1}^n \left(y_i-b-m x_i\right)}{n-2}\right]^{\frac{1}{2}} \,.
\end{equation}


Sustituyendo los valores numéricos se obtiene
$$
S_y= 0.2640 \,, \qquad \Delta R=1.412 \, \Omega \,, \qquad \Delta b= 0.06 \, V
$$

Por lo tanto, por el método de mínimos cuadrados se obtiene un valor para la resistencia de
$$
R= 150 \pm 1 \, \Omega
$$
Lo que representa un error porcentual del 0.6 %.


Teóricamente $b$ debería ser cero, pero en este trabajo se obtuvo el valor de $b= 0.07 \pm 0.06$ V.


Este resultado se puede comparar con el que se obtiene usando la librería de Numpy <<linalg.lstsq>>. Con esta librería numérica se obtiene el mismo resultado: $(m=150.449,b= 0.066)$

```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
# Los datos
Vn= array([10.120, 9.060, 8.100, 7.120, 6.140, 5.110, 4.050, 3.064, 2.137, 1.090, -10.080, -9.050, -8.050, -7.120, -6.110, -5.090, -4.120, -3.091, -2.058, -1.106])
In= array([0.067, 0.060, 0.054, 0.046, 0.041, 0.033, 0.027, 0.020, 0.013, 0.005,-0.067, -0.060, -0.055, -0.048, -0.041, -0.035, -0.027, -0.020, -0.014, -0.007])
```

```python
# El número de datos
n=len(Vn)
n
```

```python
# Minimos cuadrados
n=len(Vn)
# Las sumatorias necesarias
Sum_I= sum (In)
Sum_V= sum (Vn)
Sum_II= sum (In**2)
Sum_IV=sum (In*Vn)
print(n,',', Sum_I, ',', Sum_V, ',', Sum_II, ',', Sum_IV)
```

```python
# El cálculo de b y m
b=(Sum_II*Sum_V-Sum_IV*Sum_I)/(n*Sum_II-Sum_I**2)
m=(n*Sum_IV-Sum_I*Sum_V)/(n*Sum_II-Sum_I**2)
print('m=',m.round(3), ',', 'b=',b.round(4))
```

```python
# se hace la gráfica V vs I

# Datos
x1=In
y1=Vn

# linea recta
x2=In
y2= m*x2+b

fig,ax = plt.subplots()
# gráfica de los datos
yerr = 0.001
xerr = 0.001

ax.errorbar(x1, y1, yerr=yerr, xerr=xerr,fmt='+',color='b')
ax.set_xlabel('I [A]')
ax.set_ylabel('V [V]')

# gráfica de la recta
ax2 = ax.twinx()
ax2.plot(x2, y2, color="red", linewidth=+1, label= f'$V =$ {m:.0f} $I$ + {b:.2f}')
plt.title('Voltaje vs Corriente', fontsize=20)
plt.legend()
plt.grid(axis='both',color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
# Escribimos las sumatorias para calcular Sy
Sum_d= Sum_V -b-m*Sum_I
Sy=sqrt(Sum_d/(n-2))
print(n,',', Sum_d ,',', Sy)
```

```python
# Cálculo de los errores absolutos de m y b
Δm= (n/((n*Sum_II-Sum_I**2)))**(1/2)*Sy
Δb= (Sum_II/((n*Sum_II-Sum_I**2)))**(1/2)*Sy
print('Δm=', Δm, ',', 'Δb=',Δb)
```
